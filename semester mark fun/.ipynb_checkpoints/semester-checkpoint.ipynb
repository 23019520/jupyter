```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Subject': ['COM3120', 'COM3121', 'COM3128'],
    'Assignment Avg': [74.0, 85.5, 86.0],
    'Test Avg': [80.0, 84.0, 78.0],
    'Period Mark': [77.6, 84.6, 81.2],
    'Required Exam': [71.1, 60.6, 65.7]
}
```

```python
df = pd.DataFrame(data)
```

```python
plt.figure()
plt.bar(df['Subject'], df['Period Mark'])
plt.title('Period Marks by Subject')
plt.xlabel('Subject')
plt.ylabel('Period Mark')
plt.show()
```

```python
plt.figure()
plt.pie(df['Required Exam'], labels=df['Subject'], autopct='%1.1f%%')
plt.title('Required Exam Score Distribution')
plt.show()
```

```python
plt.figure()
plt.plot(df['Subject'], df['Assignment Avg'], marker='o', label='Assignment Avg')
plt.plot(df['Subject'], df['Test Avg'], marker='o', label='Test Avg')
plt.title('Assignment vs Test Averages')
plt.xlabel('Subject')
plt.ylabel('Average Score')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.scatter(df['Assignment Avg'], df['Test Avg'])
for i, subj in enumerate(df['Subject']):
    plt.annotate(subj, (df['Assignment Avg'][i], df['Test Avg'][i]))
plt.title('Assignment Avg vs Test Avg')
plt.xlabel('Assignment Avg')
plt.ylabel('Test Avg')
plt.show()
```

```python
plt.figure()
plt.hist(df['Test Avg'], bins=3)
plt.title('Distribution of Test Averages')
plt.xlabel('Test Avg')
plt.ylabel('Frequency')
plt.show()
```
